# Handling Missing Values

```python
df.isnull().sum() 
```

```python
updated_df = df
updated_df['retail_price']= updated_df['retail_price'].fillna(df['retail_price'].median())
updated_df.info() 
```

```python
updated_df['discounted_price']= updated_df['discounted_price'].fillna(df['discounted_price'].median())
updated_df.info() 
```

# For categorical value use forward fill(ffill) & backward fill(bfill)

```python
updated_df = df
updated_df['image'] = updated_df['image'].bfill() 
updated_df.info()
```

```python
updated_df['description'] = updated_df['description'].ffill() 
updated_df.info() 
```

```python
updated_df['brand'] = updated_df['brand'].bfill()
updated_df.info() 
```

```python
updated_df['product_specifications'] = updated_df['product_specifications'].ffill() 
updated_df.info() 
```

# Data distribution analysis
1.Distribution of retail price.

```python

plt.figure(figsize=(8,5))
sns.histplot(df['retail_price'], bins = 200, kde = True)
plt.xlabel('Frequency') 
plt.ylabel('retail_price')
plt.title('Distribution of Retail Price')
plt.show()
```

#### What is the distribution of product prices? 

```python
updated_df['retail_price'] = pd.to_numeric(df['retail_price'], errors = 'coerce')

updated_df = updated_df[updated_df['retail_price'] < updated_df['retail_price'].quantile(0.99)]

plt.figure(figsize=(10,5))
sns.histplot(updated_df['retail_price'], bins = 50, kde = True, color = 'blue')

plt.xlabel('retail_price')
plt.ylabel('Frequency')
plt.title("Distribution of product prices")

plt.show()
```

#### What is the distribution of product ratings?

```python
updated_df['overall_rating'] = pd.to_numeric(df['overall_rating'], errors = 'coerce')

updated_df = updated_df[updated_df['overall_rating'] < updated_df['overall_rating'].quantile(0.99)]

plt.figure(figsize=(10,5))
sns.histplot(updated_df['overall_rating'], bins = 50, kde = True, color = 'blue')

plt.xlabel('overall_rating')
plt.ylabel('Frequency')
plt.title("Distribution of overall rating") 

plt.show()
```

#### What are the top 10 most common product names?

```python
df1 = updated_df['product_name'].value_counts().head(10) 

plt.figure(figsize=(12,6))
sns.barplot(x= df1.index, y= df1.values)

plt.xlabel('product_name')
plt.ylabel('count') 
plt.title("Top 10 Most Common Product Name") 

plt.xticks(rotation=45)
plt.show() 
```

#### Create a heatmap to show correlations between numerical features.

```python
updated_df['retail_price'] = pd.to_numeric(updated_df['retail_price'], errors = 'coerce')

updated_df['discounted_price'] = pd.to_numeric(updated_df['discounted_price'], errors = 'coerce')

updated_df['overall_rating'] = pd.to_numeric(updated_df['overall_rating'], errors = 'coerce')

num_df = updated_df.select_dtypes(include = ['number'])

correlation_matrix = num_df.corr()

plt.figure(figure= (10, 6))
sns.heatmap(correlation_matrix , annot = True, cmap = 'coolwarm', fmt = ".2f", linewidths = 0.5)

plt.title("Correlation Heatmap of Numerical Features")

plt.show() 

```
